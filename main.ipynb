Explicació del codi:

1. from golejadors_premier_league import obtenir_golejadors_premier_league: Importa la funció del fitxer golejadors_premier_league.py.
2. Defineix els paràmetres per a any_inici, any_final, posicio, i num_futbolistes.
3. Executa la funció obtenir_golejadors_premier_league per generar el fitxer CSV.
4. Defineix el nom del fitxer generat, que inclou els paràmetres usats per a facilitar la identificació del fitxer.
5. Llegeix el contingut del fitxer CSV i mostra les dades a la consola.

```python
from golejadors_premier_league import obtenir_golejadors_premier_league
import csv

# Executar la funció amb els paràmetres desitjats
any_inici = 1993
any_final = 2024
posicio = "Mittelfeld"  # Ex. 'Mittelfeld' per migcampistes
num_futbolistes = 250

# Cridar a la funció per obtenir les dades i desar-les en un CSV
obtenir_golejadors_premier_league(any_inici, any_final, posicio, num_futbolistes)

# Nom del fitxer generat per la funció
nom_fitxer = f'golejadors_premier_league_{any_inici}_{any_final}_{posicio}_{num_futbolistes}.csv'

# Llegir i mostrar el contingut del fitxer CSV
print("Contingut del fitxer CSV:")
with open(nom_fitxer, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
